No data is clean and accurate so we need to clean our json data

as we see in the data in the d[index]["Experience"] has many startdate as null values So we need to delete the dict from the list whose startDate value is null so we won't get any problem further.

### Deleting the dict from experience list if starDate is NULL

```python
# deleting the dict from experience list if starDate is NULL

for i in range(len(d)):
    list2=d[i]["Experience"]
    if(len(d[i]['Experience'])==0):
        pass
    else:
        x=1
        for j in range(len(d[i]["Experience"])-x):
            if(d[i]['Experience'][j]['startDate']==""):
                del d[i]['Experience'][j]
                x=x+1
```

### Sorting the list of dict to date wise wrt startDate
As the experience list has many dict of experiences which are not in order so there is a function written which makes the list of dict of experiences in order with respect to date using the function - try_sorting_date()

```python
# sorting the list of dict to date wise wrt startDate

from datetime import datetime
def try_sorting_date(text,i):
    for fmt in ('%B %Y','%Y','%Y-%m-%d','%Y-%d-%m','%Y-%m','%b %Y','%Y-%m-%dT%H:%M:%S.000Z'):
        try:
            if(datetime.strptime(text, fmt)):
                d[i]["Experience"].sort(key = lambda x: datetime.strptime(x['startDate'],fmt))
        except ValueError:
            pass
```

```python
for i in range(len(d)):
    if(len(d[i]['Experience'])==0):
        pass
    else:
        j=0
        if(d[i]["Experience"][j]['startDate']==''):
            j=j+1
        try_sorting_date(d[i]["Experience"][j]['startDate'],i)
```

### Filling null values of enddate as startdate
As we see some of the endDate are null in experience section 
to remove this null enddate we filled it by the next startDate
for example- If d[0]['experience'][j]['enddate'] is null so we will put d[0]['experience'][j+1]['startdate'] which is the next joining date.

```python
# filling null values of enddate as startdate

from datetime import date
for i in range(len(d)):
    for j in range(len(d[i]["Experience"])-1):
        if(d[i]['Experience'][j]['endDate']==''):
            d[i]['Experience'][j]['endDate']=d[i]['Experience'][j+1]['startDate']
```

### Make startDate in year format and it will make the date which was in string as int
Our data set don't have all the date in same format which leads difficulties in finding the duration between them so we wrote a function called str_to_int(text,i) to make the date in same formate i.e, getting only year from it and it convert date in int format which is easier to calculate.

#### For startDate

```python
# to make startDate in year format and it will make the date which was in string as int
import datetime
def str_to_int(text,i):
    for fmt in ('%B %Y','%Y','%Y-%m-%d','%Y-%m','%b %Y','%Y-%m-%dT%H:%M:%S.000Z'):
        try:
            if(datetime.datetime.strptime(text, fmt)):
                for x in range(len(d[i]['Experience'])):
                    d[i]['Experience'][x]['startDate']=datetime.datetime.strptime(d[i]['Experience'][x]['startDate'],fmt).year
        except ValueError:
            pass
```

```python
import datetime
for i in range(len(d)):
    if(len(d[i]['Experience'])==0):
        pass
    else:
        str_to_int(d[i]['Experience'][0]['startDate'],i)
```

#### For endDate

```python
# to make endDate in year format and it will make the date which was in string as int
import datetime
def str_to_int_endDate(text,i):
    for fmt in ('%B %Y','%Y','%Y-%m-%d','%Y-%m','%b %Y','%Y-%m-%dT%H:%M:%S.000Z'):
        try:
            if(datetime.datetime.strptime(text, fmt)):
                for x in range(len(d[i]['Experience'])):
                    d[i]['Experience'][x]['endDate']=datetime.datetime.strptime(d[i]['Experience'][x]['endDate'],fmt).year
        except ValueError:
            pass
```

```python
import datetime
for i in range(len(d)):
    if(len(d[i]['Experience'])==0):
        pass
    else:
        str_to_int_endDate(d[i]['Experience'][0]['endDate'],i)
```
